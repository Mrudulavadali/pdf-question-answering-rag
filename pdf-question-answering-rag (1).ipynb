```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader(r"C:\Users\mrudu\OneDrive\Desktop\NLP (sw,token,re,).pdf")
documents = loader.load()
print(f"Loaded {len(documents)} pages")
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
docs = splitter.split_documents(documents)
print(f"Split into {len(docs)} chunks")
```

```python
from langchain_community.vectorstores import FAISS
from langchain_openai import OpenAIEmbeddings

embeddings = OpenAIEmbeddings()
db = FAISS.from_documents(docs, embeddings)
print("Vector store created!")
```

```python
!pip install langchain --upgrade
```

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model="gpt-3.5-turbo")

# Get relevant docs
docs_found = db.similarity_search("What is this document about?")
context = " ".join([d.page_content for d in docs_found])

# Ask question
response = llm.invoke(f"Based on this context: {context}\n\nWhat is this document about?")
print(response.content)
```

```python
# Ask different questions!
questions = [
    "What are the main topics?",
    "Summarize the document",
    "What is mentioned about tokens?"
]

for q in questions:
    docs_found = db.similarity_search(q)
    context = " ".join([d.page_content for d in docs_found])
    response = llm.invoke(f"Based on this context: {context}\n\n{q}")
    print(f"Q: {q}")
    print(f"A: {response.content}")
    print("---")
```
